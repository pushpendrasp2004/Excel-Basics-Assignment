1.Explain the difference between Relative , Absolute and Mixed Cell Referencing.
- Relative Cell Reference: The reference changes automatically when you copy the formula to another cell.

Example: If you write this formula in C2: =A2+B2

Copy it down to C3, it becomes: =A3+B3


- Absolute Cell Reference: The reference does not change when the formula is copied.

A dollar sign ($) is used before the column letter and row number.

- Mixed Cell Reference: Either the row or column is fixed, but not both.

One $ sign is used.
